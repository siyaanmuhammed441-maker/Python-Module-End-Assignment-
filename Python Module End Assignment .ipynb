```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("myexcel - myexcel.csv.csv")

df.head()
```

```python
np.random.seed(42)  

df['Height'] = np.random.randint(150, 181, size=len(df))

df[['Height']].head()
```

```python
team_count = df['Team'].value_counts()
team_percentage = (team_count / len(df)) * 100

team_distribution = pd.DataFrame({
    'Employee Count': team_count,
    'Percentage (%)': team_percentage
})

print(team_distribution)
```

```python
team_count.plot(kind='bar')
plt.title("Employee Distribution by Team")
plt.xlabel("Team")
plt.ylabel("Number of Employees")
plt.show()
```

```python
position_count = df['Position'].value_counts()
print(position_count)
```

```python
position_count.plot(kind='bar')
plt.title("Employee Distribution by Position")
plt.xlabel("Position")
plt.ylabel("Count")
plt.show()
```

```python
bins = [18, 25, 35, 45, 55, 65]
labels = ['18-25', '26-35', '36-45', '46-55', '56-65']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

age_group_count = df['Age Group'].value_counts()
print(age_group_count)
```

```python
age_group_count.plot(kind='bar')
plt.title("Predominant Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.show()
```

```python
team_salary = df.groupby('Team')['Salary'].sum()
print("Highest Salary by Team:")
print(team_salary.sort_values(ascending=False))
```

```python
position_salary = df.groupby('Position')['Salary'].sum()
print("\nHighest Salary by Position:")
print(position_salary.sort_values(ascending=False))
```

```python
team_salary.plot(kind='bar')
plt.title("Salary Expenditure by Team")
plt.ylabel("Total Salary")
plt.show()
```

```python
correlation = df['Age'].corr(df['Salary'])
print("Correlation between Age and Salary:", correlation)
```

```python
plt.scatter(df['Age'], df['Salary'])
plt.title("Age vs Salary Correlation")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
